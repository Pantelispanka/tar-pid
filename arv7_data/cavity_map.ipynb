```python
# dir_path = "/Users/pantelispanka/miniconda3/envs/docking-examples/cavity_finder/test/a_syn_data"

dir_path = "/Users/pantelispanka/miniconda3/envs/docking-examples/cavity_finder/test/arv7_data/arv7/from_ensemble"

dir_path_fixed = "/Users/pantelispanka/miniconda3/envs/docking-examples/cavity_finder/test/arv7_data/arv7/fixed"

# print(os.walk("/Users/pantelispanka/miniconda3/envs/docking-examples/cavity_finder/test/a_syn_data"))
```

```python
res = []

# Iterate directory
for path in os.listdir(dir_path_fixed):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        res.append(path)
print(res)
```

```python
from Bio.PDB import PDBParser, PDBIO
# from Bio.


for pdb in res:
    pdb_path = dir_path + "/" + pdb

    
    parser = PDBParser()
    structure = "Model " + pdb.split("_")[0]
    structure = parser.get_structure(structure, pdb_path)
    fixed_path = dir_path_fixed + "/" + pdb

    io = PDBIO()

    io.set_structure(structure)
    io.save(fixed_path)
```

```python
import nglview
import matplotlib as mpl
import matplotlib.pyplot as plt
import pyKVFinder
import os
import toml
import json

in_cavity = {}

for pdb in res:
    pdb_path = dir_path_fixed + "/" + pdb
    if os.path.exists(dir_path + "/results_" +  pdb.split(".")[0] + ".toml") is False:
        try:
            print("READING _PDB")
            results = pyKVFinder.run_workflow(pdb_path, include_depth=True, include_hydropathy=True, hydrophobicity_scale='EisenbergWeiss')
            print("WRITING RESULTS AT")
            print(dir_path + "/results_" +  pdb.split(".")[0] + ".toml")
            results_export = dir_path + "/results_" +  pdb.split(".")[0] + ".toml"

            cavity_export = dir_path + "/cavity_" +  pdb.split(".")[0] + ".pdb"
            barplots = dir_path + "/barplots_" +  pdb.split(".")[0] + ".pdf"
            results.export_all(fn=results_export, output=cavity_export, include_frequencies_pdf=True, pdf=barplots)
            results_file = results_export
            cavities = cavity_export
            results = toml.load(results_file)
    #         print('Volume: ', end='')
    #         print(results['RESULTS']['VOLUME'])
    #         print('Area: ', end='')
    #         print(results['RESULTS']['AREA'])
    #         print('Hydropathy: ', end='')
    #         print(results['RESULTS']['AVG_HYDROPATHY'])
            for cavity in results['RESULTS']['AVG_HYDROPATHY']:
                if results['RESULTS']['AVG_HYDROPATHY'][cavity] < 0:
    #                 print('Hydropathy: ', end='')
    #                 print(cavity)
    #                 print(results['RESULTS']['AVG_HYDROPATHY'][cavity])
    #                 print('Residues: ', end = '')
    #                 print(results['RESULTS']['RESIDUES'][cavity])
                    for residue in results['RESULTS']['RESIDUES'][cavity]:
                        resid = residue[0] + "_" + residue[2]
                        try:
                            in_cavity[resid]['in_cavities'] += 1
                            in_cavity[resid]['cavity']['cavity_name'].append(cavity)
                            in_cavity[resid]['cavity']['cavity_struct'].append(pdb)
                        except KeyError as e:
                            in_cavity[resid] = {}
                            in_cavity[resid]['in_cavities'] = 1

                            in_cavity[resid]['cavity'] = {}
                            in_cavity[resid]['cavity']['cavity_name'] = [cavity]
                            in_cavity[resid]['cavity']['cavity_struct'] = [pdb]

    #             print(in_cavity)
    #         print('Residues: ', end = '')
    #         print(results['RESULTS']['RESIDUES'])
        except Exception as e:
            print("EXCEPTIN")
            print(str(e))
            pass
    
    
with open(dir_path_fixed + "/results_all.json", "w") as outfile:
    json.dump(in_cavity, outfile)


    
# pdb = "/Users/pantelispanka/miniconda3/envs/docking-examples/cavity_finder/test/a_syn_data/182_a.pdb"
# results = pyKVFinder.run_workflow(pdb, include_depth=True, include_hydropathy=True, hydrophobicity_scale='EisenbergWeiss')

# print(results.avg_hydropathy)

# results.export_all(fn='./results.toml', output='./cavity.pdb', include_frequencies_pdf=True, pdf='./barplots.pdf')


# results_file = './results.toml'
# cavities = './cavity.pdb'



# import toml

#         # Load KVFinder results
# results = toml.load(results_file)

#         # Define function to rotate object
# def rotate_view(view, x=0, y=0, z=0, degrees=True):
#     radians = 1
#     if degrees: radians = 3.1416 / 180
#     view.control.spin([1, 0, 0], x * radians)
#     view.control.spin([0, 1, 0], y * radians)
#     view.control.spin([0, 0, 1], z * radians)

# # Visualize cavity (blue) and surface (red) points

# # Create NGL Viewer widget
# view = nglview.NGLWidget()

# # Add protein
# p1 = view.add_component(pdb)
# p1.clear()
# p1.add_cartoon(color='grey')
# # Uncomment to add transparency to cartoon representation and comment the line above
# # p1.add_cartoon(color='grey', opacity=0.5)

# # Add cavities
# p2 = view.add_component(cavities)
# p2.clear()
# p2.add_point(selection='.HA', color='red')
# p2.add_point(selection='.H', color='blue')

#         # Show volume and area
# print('Volume: ', end='')
# print(results['RESULTS']['VOLUME'])
# print('Area: ', end='')
# print(results['RESULTS']['AREA'])
# print('Residues: ', end = '')
# print(results['RESULTS']['RESIDUES'])

#         # Show
# view
```

```python
# f = open("./results_all.json")

f = open(dir_path_fixed + "/results_all.json")

# returns JSON object as 
# a dictionary
in_cavity = json.load(f)


residues = []
in_cavs = []
for key in in_cavity:
    residues.append(key)
    in_cavs.append(in_cavity[key]['in_cavities'])
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(residues, in_cavs, color ='maroon',
        width = 0.4)
 
plt.xlabel("residues")
plt.ylabel("in cavities")
plt.title("Resids in cavities")
plt.show()
```
